```python
n_years = 10
index = pd.to_datetime(np.arange(0, n_years * 365, 1), unit="D", origin="2005")
index.name = "Time [Years]"

# 1. Sine series: uniform sampling
np.random.seed(0)
data = np.sin(
    np.linspace(0, n_years * 2 * np.pi, len(index))
) 
v = pd.Series(data=data, index=index, name="STS_SIN")

# 2. Sine series: three sampling densities
v1 = pd.concat(
    [
        v.iloc[: -7 * 365].asfreq("30D"),
        v.iloc[-7 * 365 : -4 * 365].asfreq("14D"),
        v.iloc[-4 * 365 :],
    ]
)
v1.name = "STS_SIN1"

# 3. Sine series: three sampling densities + a data gap
v2 = pd.concat(
    [
        v.iloc[: -7 * 365].asfreq("30D"),
        v.iloc[-7 * 365 : -4 * 365].asfreq("14D"),
        v.iloc[-3 * 365 :],
    ]
)
v2.name = "STS_SIN2"

# 4. Sine series re-indexed onto real measurement dates
v3 = v.reindex(index_test).dropna()
v3.name = "STS_SIN3"

# Convolve white noise with an exponential kernel -> AR(1)-like series
# (the kernel IS the analytical ACF of this process; used as ground truth below)
alpha = 10
np.random.seed(0)
n = np.random.rand(index.size + 365)
b = np.exp(-np.arange(366) / alpha)
n = np.convolve(n, b, mode="valid")
n = n - n.mean()

index = pd.to_datetime(np.arange(0, n.size, 1), unit="D", origin="2005")
index.name = "Time [Years]"
n = pd.Series(data=n, index=index, name="STS_EXP")

n1 = n.reindex(v1.index).dropna()
n1.name = "STS_EXP1"
n2 = n.reindex(v2.index).dropna()
n2.name = "STS_EXP2"
n3 = n.reindex(index_test).dropna()
n3.name = "STS_EXP3"

d = pd.concat([v, v1, v2, v3,n,n1,n2,n3], axis=1)
d.plot(subplots=True, figsize=(10, 6), marker=".", markersize=2, color="steelblue");
```

```python
lags_max = 365
lags = np.arange(1.0, lags_max + 1)
```

```python
try:
    acf_df
except NameError:
    acf_df = pd.DataFrame(index=lags)

for name, sts in d.items():
    for method in ["gaussian", "rectangle"]:
        # Pastas ACF
        acf = ps.stats.acf(sts.dropna(), lags=lags_max, bin_method=method, max_gap=30)
        acf.index = acf.index.days
        acf_df.loc[:, name + "_" + method] = acf
```

```python
lags_max = 365
lags = np.arange(1.0, lags_max + 1)  # final index for PASTAS/NUFFT (without lag=0)

try:
    nufft_acf_df
except NameError:
    nufft_acf_df = pd.DataFrame(index=lags)  # lags 1..365
    # index = [1, 2, ..., 365]

for name, sts in d.items():
    print(f"{name}: n={len(sts.dropna())} points")
    sts_clean = sts.dropna()
    t = sts_clean.index.to_numpy()
    x = sts_clean.values
    t_numeric = (t - t[0]).astype("timedelta64[D]").astype(float)

    for method in ["gaussian", "rectangle"]:
        # compute_acf_*_nufft handles lag=0 and normalisation internally.
        func = compute_acf_gaussian_nufft if method == "gaussian" else compute_acf_rectangle_nufft
        nufftacf_normalized, b_nufft = func(lags=lags, t=t_numeric, x=x, bin_width=0.5)

        nufft_acf_df.loc[:, name + "_nufft_" + method] = nufftacf_normalized
```

```python
# True (analytical) autocorrelation functions
acf_v_true = pd.Series(np.cos(np.linspace(0, 2 * np.pi, 365)))

# `b` is the exponential kernel defined in the series-generation cell above;
# for an AR(1) process this kernel IS the analytical ACF.
# Requires cells to be executed in order.
acf_n_true = pd.Series(b)
```

## Plots

Each panel shows: the signal (top), ACF estimate vs ground truth (middle),
and residuals for gaussian and rectangle kernels separately (bottom).

```python
import matplotlib.gridspec as gridspec
from matplotlib.dates import YearLocator, DateFormatter

def plot_acf_val_err(
    names=["STS_EXP1", "STS_EXP2", "STS_EXP3"],
    acf_true=acf_n_true,
    logx=False):

  fig = plt.figure(figsize=(15, 10))

  outer = gridspec.GridSpec(2, 1, height_ratios=[1, 2.2], hspace=0.1, figure=fig)

  gs_signal = outer[0].subgridspec(1, 3, wspace=0.15)

  gs_acf = outer[1].subgridspec(3, 3, hspace=0.0, wspace=0.2)

  axes_signal = [fig.add_subplot(gs_signal[0, j]) for j in range(3)]

  axes_acf = np.empty((3, 3), dtype=object)
  for j in range(3):
      ax0 = fig.add_subplot(gs_acf[0, j], sharey=axes_acf[0, 0] if j > 0 else None)
      ax1 = fig.add_subplot(gs_acf[1, j], sharex=ax0, sharey=axes_acf[1, 0] if j > 0 else None)
      ax2 = fig.add_subplot(gs_acf[2, j], sharex=ax0, sharey=axes_acf[2, 0] if j > 0 else None)
      axes_acf[0, j], axes_acf[1, j], axes_acf[2, j] = ax0, ax1, ax2

  for j, name in enumerate(names):
      # --- Signal ---
      sts = d.loc[:, name]
      sts.plot(ax=axes_signal[j], style=".", legend=True, label=name)
      axes_signal[j].xaxis.set_major_locator(YearLocator(base=3))
      axes_signal[j].xaxis.set_major_formatter(DateFormatter('%Y'))

      ax_val   = axes_acf[0, j]   # ACF : truth + gaussian
      ax_err_g = axes_acf[1, j]   # ACF gaussian residual
      ax_err_r = axes_acf[2, j]   # AXF rectangle residual

      # ================= GAUSSIAN kernel =================
      acf = acf_df.loc[:, f"{name}_gaussian"]
      pastas_kwargs = dict(marker="x", markersize=10, color="blue", linestyle="",logx=logx)
      acf.plot(label="PASTAS", ax=ax_val, **pastas_kwargs)

      dif = acf.subtract(acf_true).dropna()
      rmse = f" (RMSE={np.sqrt((dif.pow(2)).sum() / dif.size):.2f})"
      dif.plot(label=f"PASTAS {rmse}", ax=ax_err_g, **pastas_kwargs)

      nufft_name = f"{name}_nufft_gaussian"
      nufft_kwargs = dict(linestyle="-", color="red", linewidth=1.5, marker=".", markersize=6,logx=logx)
      if nufft_name in nufft_acf_df.columns:
          nufftacf = nufft_acf_df.loc[:, nufft_name]
          nufftacf.plot(label="NUFFT", ax=ax_val, **nufft_kwargs)

          dif_nufft = nufftacf.subtract(acf_true).dropna()
          rmse_n = f" (RMSE={np.sqrt((dif_nufft.pow(2)).sum() / dif_nufft.size):.2f})"
          dif_nufft.plot(label=f"NUFFT {rmse_n}", ax=ax_err_g, **nufft_kwargs)

      ax_err_g.axhline(0, color="black", linewidth=0.8, linestyle=":")

      # ================= RECTANGLE kernel (only residual ) =================
      acf = acf_df.loc[:, f"{name}_rectangle"]
      pastas_kwargs = dict(marker="x", markersize=10, color="blue", linestyle="",logx=logx)

      dif = acf.subtract(acf_true).dropna()
      rmse = f" (RMSE={np.sqrt((dif.pow(2)).sum() / dif.size):.2f})"
      dif.plot(label=f"PASTAS {rmse}", ax=ax_err_r, **pastas_kwargs)

      nufft_name = f"{name}_nufft_rectangle"
      nufft_kwargs = dict(linestyle="-", color="red", linewidth=1.5, marker=".", markersize=6,logx=logx)

      if nufft_name in nufft_acf_df.columns:
          nufftacf = nufft_acf_df.loc[:, nufft_name]
          dif_nufft = nufftacf.subtract(acf_true).dropna()
          rmse_n = f" (RMSE={np.sqrt((dif_nufft.pow(2)).sum() / dif_nufft.size):.2f})"
          dif_nufft.plot(label=f"NUFFT {rmse_n}", ax=ax_err_r, **nufft_kwargs)

      ax_err_r.axhline(0, color="black", linewidth=0.8, linestyle=":")

      # --- ACF truth ---
      acf_true.plot(ax=ax_val, color="black", linestyle="--", linewidth=1, label="true ACF")

      ax_err_r.set_xticks([1, 10, 30, 60, 180, 365])
      ax_err_r.set_xticklabels([1, 10, 30, 60, 180, 365])
      ax_err_r.set_xlabel("Lags [Days]")

      ax_val.tick_params(labelbottom=False)
      ax_err_g.tick_params(labelbottom=False)

      ax_val.legend(loc="lower left")
      ax_err_g.legend(loc="lower left")
      ax_err_r.legend( loc="lower left")

  axes_signal[0].set_ylabel("Signal")
  axes_acf[0, 0].set_ylabel("ACF gauss")
  axes_acf[1, 0].set_ylabel("Err. gauss.")
  axes_acf[2, 0].set_ylabel("Err. rect.")
  for j in range(1, 3):
      for r in range(3):
          axes_acf[r, j].set_ylabel("")

  plt.show()
```
